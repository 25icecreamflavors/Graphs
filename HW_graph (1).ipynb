1. Given the network, give a short description what is it about.

I took bio-diseasome network. It is a network of disorders and disease genes linked by known disorder–gene associations. Genes associated with similar disorders show both higher likelihood of physical interactions between their products and higher expression profiling similarity for their transcripts, supporting the existence of distinct disease-specific functional modules.

Authors constructed a bipartite graph consisting of two disjoint sets of nodes. One set corresponds to all known genetic disorders, whereas the other set corresponds to all known disease genes in the human genome. A disorder and a gene are then connected by a link if mutations in that gene are implicated in that disorder

In the “human disease network” (HDN) nodes represent disorders, and two disorders are connected to each other if they share at least one gene in which mutations are associated with both disorders. In the “disease gene network” (DGN) nodes represent disease genes, and two genes are connected if they are associated with the same disorder.

### 2.1. Check that Zipf's Law holds <br>

Get zipf law values for next graph

```python
# degree distribution
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
```

```python
check_list = []
first_freq = degree_sequence[0]

for i in range(1, len(degree_sequence) + 1):
    freq = first_freq / i
    check_list.append(freq)
```

```python
fig = plt.figure(figsize=(14,7))

plt.title("Degree count vs Zipf's law plot", size=18)
plt.xlabel("Position", size=14)
plt.ylabel("Degree", size=14)
plt.plot(check_list, label="Zipf's Law")
plt.plot(degree_sequence, label="Real frequency")
plt.legend()
plt.show()
```

We actually see that zipf's law doesn't really work here, graphs are slightly similiar but not so much.

### 2.3. Print 5 nodes with maximum degree centrality, vizualize your graph with color or size as the function of degree <br>

```python
from collections import Counter
```

```python
temp_dict = nx.degree_centrality(G)
print(dict(Counter(temp_dict).most_common(5)))
```

```python
plt.figure(figsize=(15,10))

deg_centrality = nx.degree_centrality(G)
centrality = np.fromiter(deg_centrality.values(), float)
sizes = centrality  / np.max(centrality) * 200
normalize = mcolors.Normalize(vmin=centrality.min(), vmax=centrality.max())
colormap = cm.viridis

scalarmappaple = cm.ScalarMappable(norm=normalize, cmap=colormap)
scalarmappaple.set_array(centrality)

plt.colorbar(scalarmappaple)

# plot
pos = nx.kamada_kawai_layout(G)
nx.draw_networkx_edges(G, pos, alpha=0.2, edge_color='#25283D')
nx.draw_networkx_nodes(G, pos, node_color=sizes, node_size=sizes,
                       cmap=colormap)
plt.title("Degree centralities on the graph", size=18)
plt.show()
```

### 3.5. Implement community detection using algorithm from the table

We will choose 7 clusters, because there were about 7 communities in the previous algorithm and it looked good.

```python
X = model[model.wv.vocab]
```

```python
from nltk.cluster import KMeansClusterer
import nltk

NUM_CLUSTERS=7
kclusterer = KMeansClusterer(NUM_CLUSTERS, distance=nltk.cluster.util.cosine_distance, repeats=25)
assigned_clusters = kclusterer.cluster(X, assign_clusters=True)
print(assigned_clusters)
```

### 3.6. Perform dimension reduction and visualize resulting clusters with matplotlib

```python
clust_dict = {}

nodes = list(model.wv.vocab)
for i, node in enumerate(nodes):  
    clust_dict[node] = assigned_clusters[i]
```

```python
plt.figure(figsize=(15,10))

# plot
nodes = G.nodes()
colors = [clust_dict[str(i)] for i in range(len(G.nodes))]

pos = nx.kamada_kawai_layout(G)
nx.draw_networkx_edges(G, pos, alpha=0.2, edge_color='#25283D')
nx.draw_networkx_nodes(G, pos, node_color=colors, node_size=35)
plt.title("Degree centralities on the graph", size=18)
plt.show()
```
