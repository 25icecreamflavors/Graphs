### 2.1. Check that Zipf's Law holds <br>

Get zipf law values for next graph

```python
# degree distribution
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
```

```python
check_list = []
first_freq = degree_sequence[0]

for i in range(1, len(degree_sequence) + 1):
    freq = first_freq / i
    check_list.append(freq)
```

```python
fig = plt.figure(figsize=(14,7))

plt.title("Degree count vs Zipf's law plot", size=18)
plt.xlabel("Position", size=14)
plt.ylabel("Degree", size=14)
plt.plot(check_list, label="Zipf's Law")
plt.plot(degree_sequence, label="Real frequency")
plt.legend()
plt.show()
```

We actually see that zipf's law doesn't really work here, graphs are slightly similiar but not so much.

### 2.3. Print 5 nodes with maximum degree centrality, vizualize your graph with color or size as the function of degree <br>

```python
from collections import Counter
```

```python
temp_dict = nx.degree_centrality(G)
dict(Counter(temp_dict).most_common(5))
```

```python
plt.figure(figsize=(15,10))

deg_centrality = nx.degree_centrality(G)
centrality = np.fromiter(deg_centrality.values(), float)
sizes = centrality  / np.max(centrality) * 200
normalize = mcolors.Normalize(vmin=centrality.min(), vmax=centrality.max())
colormap = cm.viridis

scalarmappaple = cm.ScalarMappable(norm=normalize, cmap=colormap)
scalarmappaple.set_array(centrality)

plt.colorbar(scalarmappaple)

# plot
pos = nx.kamada_kawai_layout(G)
nx.draw_networkx_edges(G, pos, alpha=0.2, edge_color='#25283D')
nx.draw_networkx_nodes(G, pos, node_color=sizes, node_size=sizes,
                       cmap=colormap)
plt.title("Degree centralities on the graph", size=18)
plt.show()
```
